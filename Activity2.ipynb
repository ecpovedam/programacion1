### 2.4 Limpieza de texto

Aplicar las siguientes funciones para realizar limpieza de las vaariables string:
**Strip** : Suprime blancos u otro carácter especificado del final o del principio de una expresión de serie
**Upper**:  Convierte todos los caracteres alfabéticos de la cadena en mayúscula. Todos los caracteres no alfabéticos no sufren cambios
**Replace**: Reemplaza carácteres de una cadena de texo [Tip: **\\w** palabras **\\d** digitos **\\s** espacios **\\.** puntos **^** antitésis]


```python
# Definir una función limpieza_texto para aplicar varias columnas

def limpieza_texto (Df5: pd.DataFrame, column_name:object):
    Df5[column_name]= Df5[column_name].str.strip()
    Df5[column_name]= Df5[column_name].str.upper()
    Df5[column_name]= Df5[column_name].replace("  ","")
    Df5[column_name]= Df5[column_name].replace('[^\\w\\d\\s\\.]','',regex=True)
    Df5[column_name]= Df5[column_name].replace('$','')
    Df5[column_name]= Df5[column_name].replace(['Á','É','Í','Ó','Ú'],['A','E','I','O','U'])
    return Df5
```

```python
#Aplicar la función limpieza_texto a todas las columnas del Df
limpieza_texto (Df5,'RAZÓN SOCIAL')
limpieza_texto (Df5,'SUPERVISOR')
limpieza_texto (Df5,'REGIÓN')
limpieza_texto (Df5,'DEPARTAMENTO DOMICILIO')
limpieza_texto (Df5,'CIUDAD FINAL')
limpieza_texto (Df5,'MACROSECTOR')
limpieza_texto (Df5,'INGRESOS OPERACIONALES')
limpieza_texto (Df5,'GANANCIA (PÉRDIDA)')
limpieza_texto (Df5,'TOTAL ACTIVOS')
limpieza_texto (Df5,'TOTAL PASIVOS')
limpieza_texto (Df5,'TOTAL PATRIMONIO')
Df5.head()
```

### 2.5 Limpieza númerica


```python
#Crear una función para limpieza de las variables númericas aplicando replace ()
def limpieza_numerica (Df5: pd.DataFrame, column_name:int):
    Df5[column_name]= Df5[column_name].replace("  ","")
    Df5[column_name]= Df5[column_name].replace('[^\\d\\.]','',regex=True)
    return Df5

limpieza_numerica (Df5,'NIT')
limpieza_numerica (Df5,'CIIU')
limpieza_numerica (Df5,'Año de Corte')
Df5.head()
```
