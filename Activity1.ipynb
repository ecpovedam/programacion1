**Conditions and Conditional**

```python
#mujeres que sobrevivieron
female_survives = new_df[(new_df['Sex']=="female") & (new_df['Survived']==1)]
female_survives.shape
```

```python
#hombres que sobrevivieron
male_survives = new_df[(new_df['Sex']=="male") & (new_df['Survived']==1)]
male_survives.shape
```

```python
if male_survives.shape<female_survives.shape:
    print("Las mujeres tuvieron mayor probabilidad de sobrevivir")
```
